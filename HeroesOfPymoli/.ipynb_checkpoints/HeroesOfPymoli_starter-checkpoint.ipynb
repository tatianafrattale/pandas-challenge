```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
pymoli_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
pymoli_data_df = pd.read_csv(pymoli_data)

# Print df
pymoli_data_df
```

## Player Count

* Display the total number of players


```python
# Calculate
total_players = len(pymoli_data_df["SN"].unique())

# Create Data Frame
total_players_df = pd.DataFrame({"Total Number of Players": [total_players]})

# Print df
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculations
unique_items = len(pymoli_data_df["Item ID"].unique())
average_price = pymoli_data_df["Price"].mean()
purchases_count = pymoli_data_df["Purchase ID"].count()
total_revenue = pymoli_data_df["Price"].sum() 

# Create summary data frame
total_purchasing_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                                         "Average Price": [average_price],
                                         "Total Purchases": [purchases_count],
                                          "Total Revenue": [total_revenue]})

# Cleaning up formatting to add dollar signs, decimals, and commas
total_purchasing_analysis["Average Price"] = total_purchasing_analysis["Average Price"].astype(float).map(
    "${:,.2f}".format)
total_purchasing_analysis["Total Revenue"] = total_purchasing_analysis["Total Revenue"].astype(float).map(
    "${:,.2f}".format)

# Print df
total_purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Group data by gender
gender_data = pymoli_data_df.groupby("Gender")

# Get unique gender counts by counting by unique "SN" to avoid players who were repeat customers
gender_counts = gender_data.nunique()["SN"]

# Get percentage
gender_percentages = gender_counts / total_players * 100

# Create Data Frame
gender_demo_df = pd.DataFrame({"Total Count": gender_counts, 
                               "Percentage of Players": gender_percentages})

# Round the percent to 2 decimal places and add percent sign
gender_demo_df["Percentage of Players"] = gender_demo_df["Percentage of Players"].astype(float).map(
    "{0:.2f}%".format)

# Print Data Frame
gender_demo_df


```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculations
purchase_count = gender_data["Purchase ID"].count()
av_purch_price = gender_data["Price"].mean()
total_purch = gender_data["Price"].sum()
av_purch_per_person = total_purch / gender_counts

# Creating data frame
gender_purch_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                         "Average Purchase Price": av_purch_price,
                                         "Total Purchases": total_purch,
                                          "Average Total Purchase Per Person": av_purch_per_person})

# Changing format for monetary values
gender_purch_analysis["Average Purchase Price"] = gender_purch_analysis["Average Purchase Price"].astype(float).map(
    "${:,.2f}".format)
gender_purch_analysis["Total Purchases"] = gender_purch_analysis["Total Purchases"].astype(float).map(
    "${:,.2f}".format)
gender_purch_analysis["Average Total Purchase Per Person"] = gender_purch_analysis["Average Total Purchase Per Person"].astype(float).map(
    "${:,.2f}".format)


# Print data frame
gender_purch_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Find out what min and max ages are 
print(pymoli_data_df["Age"].max())
print(pymoli_data_df["Age"].min())
```

```python
# Create bins
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 50]

# Create labels for bins
age_bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice data and place into bins
pd.cut(pymoli_data_df["Age"], bins, labels = age_bin_labels).head()
```

```python
# Place this data into a new column inside the data frame
pymoli_data_df["Age Group"] = pd.cut(pymoli_data_df["Age"], bins, labels= age_bin_labels)
pymoli_data_df.head()
```

```python
# Group data frame by age group
age_data = pymoli_data_df.groupby("Age Group")

# Find out how many players are in each bin
age_total_players = age_data.nunique()["SN"]
print(age_total_players)

# Calculate percentage of players
age_percent = age_total_players / total_players * 100
print(age_percent)
```

```python
# Create data frame
age_demo_df = pd.DataFrame({"Total Count": age_total_players,
                            "Percentage of Players": age_percent})

# Cleaning format
age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].astype(float).map(
    "{0:.2f}%".format)

# Print data frame
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin purchase data df by age
pymoli_data_df["Age Group"] = pd.cut(pymoli_data_df["Age"], bins, labels= age_bin_labels)

# Group by age group
age_data = pymoli_data_df.groupby("Age Group")

# Calculations
age_purch_count = age_data["Purchase ID"].count()
age_av_price = age_data["Price"].mean()
age_total_value = age_data["Price"].sum()
age_av_total_per_person = age_total_value / age_total_players
```

```python
# Create data frame
age_purch_analysis = pd.DataFrame({"Purchase Count": age_purch_count,
                                    "Average Purchase Price": age_av_price,
                                    "Total Purchase Value": age_total_value,
                                    "Average Total Purchase Per Person": age_av_total_per_person})
# Cleaning format
age_purch_analysis["Average Purchase Price"] = age_purch_analysis["Average Purchase Price"].astype(float).map(
    "${:,.2f}".format)
age_purch_analysis["Total Purchase Value"] = age_purch_analysis["Total Purchase Value"].astype(float).map(
    "${:,.2f}".format)
age_purch_analysis["Average Total Purchase Per Person"] = age_purch_analysis["Average Total Purchase Per Person"].astype(float).map(
    "${:,.2f}".format)

# Print data frame
age_purch_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group data by "SN" 
sn_data = pymoli_data_df.groupby("SN")

# Calculations
sn_purch_count = sn_data["Price"].count()
sn_av_price = sn_data["Price"].mean()
sn_total_purch = sn_data["Price"].sum()

top_spenders_df = pd.DataFrame({"Purchase Count": sn_purch_count,
                                "Average Purchase Price": sn_av_price,
                                "Total Purchase Value": sn_total_purch})

# Cleaning format
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].astype(float).map(
    "${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].astype(float).map(
    "${:,.2f}".format)

# Sort "Total Purchase Value" colum in descending order
top_spend_sorted = top_spenders_df.sort_values("Total Purchase Value", ascending = False)



# Printing top five in data frame
top_spend_sorted.head()

```
